```python
import seaborn as sns
sns.set_theme(style="darkgrid")
import pandas as pd
df = pd.read_csv("DSPyVanillaJac.csv")
dspySum = df["DSPyCOTExactMatch"].sum()
jacSum = df["JacExactMatch"].sum()
total = df.shape[0]
print(dspySum / total, jacSum / total, total)
plot = {"DSPyCOT": dspySum/total, "Jac": jacSum/total}
sns.barplot(plot)
```

```python
import seaborn as sns
sns.set_theme(style="darkgrid")
import pandas as pd
df = pd.read_csv("DSPyVanillaJacGSM8k.csv")
dspySum = df["DSPyCOTExactMatch"].sum()
jacSum = df["JacExactMatch"].sum()
dspyFailSum = df["DSPyCOTFailed"].sum()
jacFailSum = df["JacFailed"].sum()
total = df.shape[0]
print(dspySum / total, jacSum / total, dspyFailSum / total, total)
plot = {"DSPyCOT": dspySum/total, "Jac": jacSum/total}
sns.barplot(plot)
plot = {"DSPyCOTFailure": dspyFailSum/total, "JacFailure": jacFailSum/total}
sns.barplot(plot)
```
